Set some constants.

```python
# Directory containing our segmentation files
DATA_DIR = '/Users/ryanshaw/Code/u-series-segmentations'

# Boundary similarity parameter: Maximum distance (in potential boundary positions) that a transposition may span
MAX_TRANSPOSE_DISTANCE = 8

# Some nice colors
ALMOST_BLACK = '#262626'
DARK_GRAY = np.array([float(200)/float(255)]*3)
LIGHT_GRAY = np.array([float(248)/float(255)]*3)
```

Define some utility functions.

```python
def datapath(path):
    return os.path.join(DATA_DIR, path)

def remove_coder(dataset, coder):
    for item, coder_segmentations in dataset.items():
        dataset[item].pop(coder)
    dataset.coders.discard(coder)
    return dataset

def load_dataset(name):
    dataset = se.input_linear_mass_json(datapath(name))
    dataset.coders = set.intersection(*[ set(data.keys()) for data in dataset.values() ])
    return dataset

def micro_mean(ratios):
    num, den = [ sum(values) for values in zip(*ratios) ]
    return num / den    

def micro_mean_std_n(ratios):
    mean = micro_mean(ratios)
    scores = [ (n/d) for n,d in ratios ]
    squared_deviations = [ abs(s - mean)**2 for s in scores ]
    std = np.sqrt(np.mean(squared_deviations))
    return (mean, std, len(ratios))

def micro_mean_std_n_str(ratios):
    return '{0:.4f} ±{1:.4f}, n={2}'.format(*micro_mean_std_n(ratios))

def mean_std_n(values):
    floats = np.array(list(values), dtype=float)
    return (np.mean(floats), np.std(floats), len(floats))

def mean_std_n_str(values):
    return '{0:.4f} ±{1:.4f}, n={2}'.format(*mean_std_n(values))

def b_statistics(dataset):
    return se.boundary_statistics(dataset, n_t=MAX_TRANSPOSE_DISTANCE)

def b(*datasets, return_parts=False):
    return se.boundary_similarity(*datasets, n_t=MAX_TRANSPOSE_DISTANCE, return_parts=return_parts)

def wd(dataset):
    return se.window_diff(dataset)

def actual_agreement(dataset, return_parts=False):
    return se.actual_agreement_linear(dataset, n_t=MAX_TRANSPOSE_DISTANCE, return_parts=return_parts)

def pi(dataset, return_parts=False):
    return se.fleiss_pi_linear(dataset, n_t=MAX_TRANSPOSE_DISTANCE, return_parts=return_parts)

def kappa(dataset):
    return se.fleiss_kappa_linear(dataset, n_t=MAX_TRANSPOSE_DISTANCE)

def bias(dataset):
    return se.artstein_poesio_bias_linear(dataset, n_t=MAX_TRANSPOSE_DISTANCE)

def pairwise_similarity(dataset, coder, item):
    ratios = [ v[:2] for k,v in b(dataset, return_parts=True).items() 
               if (coder in k) and (item in k) ]
    return micro_mean(ratios)
    
def strip_prefixes(strings):
    return [ s.split(':')[-1] for s in strings ]

def merge_datasets(*datasets):
    items = set.intersection(*[ set(dataset.keys()) for dataset in datasets ])
    return se.data.Dataset(
        { item: { c:s for c,s in it.chain.from_iterable([ d[item].items() for d in datasets ]) }
          for item in items })

def total_mass(dataset, item):
    sums = [ sum(s) for s in dataset[item].values() ]
    assert(all(s == sums[0] for s in sums))
    return sums[0]

def total_masses_for(dataset):
    return { item: total_mass(dataset, item) for item in dataset.keys() }

def total_boundaries_laid(dataset, item):
    return sum([ (len(s) - 1) for s in dataset[item].values() ])

def total_segments_created(dataset, item):
    return sum([ len(s) for s in dataset[item].values() ])

def boundaries_placed(dataset, coder=None):
    if coder is not None:
        return sum([ (len(v[coder]) - 1) for v in dataset.values() ])
    else:
        return sum([ boundaries_placed(dataset, c) for c in dataset.coders ])

def potential_boundaries(dataset):
    return sum([ (total_mass(dataset, item) - 1) for item in dataset.keys() ])

def boundary_placement_rate(dataset, coder):
    return boundaries_placed(dataset, coder) / potential_boundaries(dataset)

def boundary_ratios_for(dataset):
    d = collections.defaultdict(dict)
    for item, segmentations in dataset.items():
        for coder, masses in segmentations.items():
            actual_boundaries = len(masses) - 1
            possible_boundaries = sum(masses) - 1
            d[item][coder] = (actual_boundaries, possible_boundaries)
    return d

def null_segmentations_for(dataset):
    return se.data.Dataset({ item: { 'NULL': (total_mass(dataset, item),) } 
                             for item in dataset.keys() })

def random_segmentation_of(dataset, item, ratio):
    mass = 0
    masses = []
    for x in range(total_mass(dataset, item)):
        mass += 1
        if (random.random() < ratio):
            masses.append(mass)
            mass = 0
    masses.append(mass)
    return masses

def random_segmentations_for(dataset):
    d = collections.defaultdict(dict)
    boundary_ratios = boundary_ratios_for(dataset)
    for item, coder_ratios in boundary_ratios.items():
        p_of_b = micro_mean(coder_ratios.values())
        d[item]['RANDOM'] = random_segmentation_of(dataset, item, p_of_b)
    return se.data.Dataset(d)

def regular_segmentation_of(dataset, item, num_segments):
    total_m = total_mass(dataset, item)
    segment_m = math.floor(total_m / num_segments)
    remainder = total_m % num_segments
    masses = [segment_m] * num_segments
    for i in range(remainder):
        masses[i] += 1
    return masses

def regular_segmentations_for(dataset):
    d = collections.defaultdict(dict)
    for item, coder_segmentations in dataset.items():
        num_segments = int(round(np.mean(np.array([ len(s) for s in coder_segmentations.values() ]))))
        d[item]['REGULAR'] = regular_segmentation_of(dataset, item, num_segments)
    return se.data.Dataset(d)

def title_to_filename(title):
    return title.lower().replace(' ', '-')

# Draw a Cleveland-style dot plot with possibly multiple values per label
def dot_plot(values, labels, title, value_labels=None, minv=None, draw_mean_std=False):
    if len(values) == 0: return
    
    if isinstance(values[0], (list, tuple)):
        draw_mean_std = False
    else:
        values = [ values ]

    draw_legend = True
    
    if not value_labels:
        value_labels = [''] * len(labels)
        draw_legend = False
        
    if len(values) > 1:
        from brewer2mpl import qualitative
        colors = qualitative.Set2[len(values)].mpl_colors
    else:
        colors = [ ALMOST_BLACK ]
    
    # order by mean of the multiple values
    means = [ np.mean(np.array(list(v), dtype=float)) for v in zip(*values) ]
    ordered = list(zip(*sorted(zip(*[means] + [labels] + values))))
    labels, values = ordered[1], ordered[2:]
    y = range(len(labels))

    if minv is None:
        minv = float(min(it.chain(*values)))
    maxv = float(max(it.chain(*values)))
    width = maxv - minv
    left = minv - 0.05 * width
    right = maxv + 0.05 * width

    plt.xlim(left, right)
    plt.ylim(-1, len(labels))
    
    plt.yticks(y, labels);
    plt.tick_params(axis='x', direction='in', colors=ALMOST_BLACK)
    plt.tick_params(axis='y', left='off', right='off', colors=ALMOST_BLACK)
    
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
    
    plt.hlines(y, left, right, linestyles='dotted', linewidths=0.5, colors=ALMOST_BLACK)
    plt.xlabel(title, color=ALMOST_BLACK)
    
    for v, color, label in zip(values, colors, value_labels):
        plt.plot(v, y, 'o', color=color, alpha=0.75, label=label)
        
    if draw_mean_std:
        mean, std, n = mean_std_n(values)
        plt.vlines(mean, 0, len(labels)-1, colors=[ALMOST_BLACK])
        plt.vlines([mean-std, mean+std], 0, len(labels)-1, colors=[DARK_GRAY])
        
    if draw_legend:
        leg = plt.legend(frameon=True, numpoints=1, loc='center', bbox_to_anchor=(1,0.5))
        plt.setp(leg.get_frame(), facecolor=LIGHT_GRAY, edgecolor='none')
        for t in leg.texts: t.set_color(ALMOST_BLACK)
    
    plt.savefig(title_to_filename(title) + '.svg')
    plt.show()

# Draw a box-and-whisker plot
def box_plot(values, labels, title):

    medians = [ np.median(np.array(v, dtype=float)) for v in values ]
    medians, values, labels = zip(*sorted(zip(medians, values, labels)))
    bp = plt.boxplot(values, sym='', vert=False, widths=0.7)

    plt.setp(bp['boxes'], color=ALMOST_BLACK, linewidth=0.5)
    plt.setp(bp['whiskers'], color=ALMOST_BLACK, linewidth=0.5)
    plt.setp(bp['medians'], color='#e41a1c', linewidth=0.5)

    y = range(1, len(values)+1)
    plt.ylim(0, len(values)+1)

    plt.yticks(y, labels)
    plt.tick_params(axis='x', direction='in', colors=ALMOST_BLACK)
    plt.tick_params(axis='y', left='off', right='off', colors=ALMOST_BLACK)

    for spine in plt.gca().spines.values():
        spine.set_visible(False)
    
    plt.xlabel(title, color=ALMOST_BLACK)

    plt.savefig(title_to_filename(title) + '.svg')
    plt.show()
    
def histogram(values, bins, title):
    plt.hist(values, bins=bins, color=DARK_GRAY, edgecolor=DARK_GRAY)
    plt.xlabel(title, color=ALMOST_BLACK)
    plt.savefig(title_to_filename(title) + '.svg')
    plt.show()

def heatmap(values, row_labels, column_labels, title, scale_w=0.8, scale_h=2.0):
    array = np.array(values, dtype=float)
    
    assert(array.shape[1] == len(column_labels))
    assert(array.shape[0] == len(row_labels))
    
    # Thanks https://stackoverflow.com/questions/14391959/heatmap-in-matplotlib-with-pcolor
    fig, ax = plt.subplots()
    collection = ax.pcolor(array, vmin=0, vmax=1, cmap=plt.cm.binary, alpha=0.8)
    fig.colorbar(collection)
    
    # put the major ticks at the middle of each cell
    ax.set_xticks(np.arange(array.shape[1])+0.5, minor=False)
    ax.set_yticks(np.arange(array.shape[0])+0.5, minor=False)

    # want a more natural, table-like display
    ax.invert_yaxis()
    ax.xaxis.tick_top()

    plt.xlabel(title, color=ALMOST_BLACK)
    ax.set_xticklabels(column_labels, minor=False)
    ax.set_yticklabels(row_labels, minor=False)
    
    w,h = fig.get_size_inches()
    plt.gcf().set_size_inches([w*scale_w, h*scale_h])
    plt.ylim(len(row_labels), 0)
    plt.tick_params(axis='x', top='off', bottom='off', colors=ALMOST_BLACK)
    plt.tick_params(axis='y', left='off', right='off', colors=ALMOST_BLACK)
    
    plt.savefig(title_to_filename(title) + '.svg')
    plt.show()

def table(rows, header_rows=1):
    thead = '<thead>%s</thead>' % ''.join([ '<tr>%s</tr>' % ''.join(
        [ '<th>%s</th>' % value for value in row ])
            for row in rows[:header_rows] ])
    tbody = '<tbody>%s</tbody>' % ''.join([ '<tr>%s</tr>' % ''.join(
        [ '<td>%s</td>' % value for value in row ])
            for row in rows[header_rows:] ])
    sums = [ sum(values) if (type(values[0]) == int) else '' for values in zip(*rows[header_rows:]) ]
    tfoot = '<tfoot><tr>%s</tr></tfoot>' % ''.join(
        [ '<th>%s</th>' % s for s in sums ])
    return '<table>%s</table>' % ''.join([thead,tbody,tfoot])
```

```python
def summarize_segmentations(dataset):
    head = [ [ 'Interview', 'Mass', 'PBs', 'Boundaries', 'Segments', 'Coders', 'Mean Segment Mass', 'Mean Segment Count', 'Mean Boundaries Per Coder' ] ]
    rows = sorted([ [ item.split(':')[-1],
               total_mass(dataset, item),
               total_mass(dataset, item) - 1,
               total_boundaries_laid(dataset, item),
               total_segments_created(dataset, item),
               len(coders),
               mean_std_n_str(it.chain.from_iterable(dataset[item].values())),
               mean_std_n_str([ len(s) for s in dataset[item].values() ]),
               mean_std_n_str([ (len(s) - 1) for s in dataset[item].values() ]) ]
             for item, codings in dataset.items() ], key=lambda x: x[1], reverse=True)
    return table(head + rows)

HTML(summarize_segmentations(human_segmentations))
```

```python
def gather(statistics, item, field):
    return sum([ (len(v[field]) if type(v[field]) == list else v[field])  
                 for k,v in statistics.items() if k.startswith(item) ])

def summarize_comparisons(dataset):
    stats = b_statistics(dataset)
    head = [ [ 'Interview', 'Comparisons', 'Boundaries', 'P(b)', 'Additions', 'P(a|b)', 'Transpositions', 'P(t|b)', 'Matches', 'P(m|b)' ] ]
    rows = sorted([ [ item.split(':')[-1],
               gather(stats, item, 'pbs'),
               gather(stats, item, 'boundaries_all'),
               round(gather(stats, item, 'boundaries_all') / gather(stats, item, 'pbs'), 4),
               gather(stats, item, 'additions'),
               round(gather(stats, item, 'additions') / gather(stats, item, 'boundaries_all'), 4),
               gather(stats, item, 'transpositions'),
               round(gather(stats, item, 'transpositions') / gather(stats, item, 'boundaries_all'), 4),
               gather(stats, item, 'matches'),
               round(gather(stats, item, 'matches') / gather(stats, item, 'boundaries_all'), 4) ]
             for item, codings in dataset.items() ], key=lambda x: x[1], reverse=True)
    return table(head + rows)

HTML(summarize_comparisons(human_segmentations))
```

```python
def get_sentence_lengths(interview_id):
    with open(datapath('interview-sentences/%s.txt' % interview_id)) as f:
        return [ len(l) for l in f.readlines() ]
    
sentence_lengths = [ get_sentence_lengths(id) for id in interview_ids ]
box_plot(sentence_lengths, interview_ids, 'Sentence length in characters')
print('mean sentence length: %s' % mean_std_n_str(list(it.chain(*sentence_lengths))))
```

```python
def ignore_interviewer_turns(speaker_masses):
    speaker_a = speaker_masses[::2]
    speaker_b = speaker_masses[1::2]
    if sum(speaker_a) > sum(speaker_b):
        return speaker_a
    else:
        return speaker_b
 
interviewee_masses = [ ignore_interviewer_turns(speakers['interviews:%s' % id]['speakers']) for id in interview_ids ]
box_plot(interviewee_masses, interview_ids, 'Interviewee turn length in sentences')
print('mean interviewee turn length in sentences: {0} (sd: {1})'.format(*mean_std_n(list(it.chain(*interviewee_masses)))))
```

```python
def show_similarity_and_agreement(dataset):
    ratios = [ tup[:2] for tup in b(dataset, return_parts=True).values() ]
    print('Microaveraged B \t {}'.format(micro_mean_std_n_str(ratios)))
    print('π* (Fleiss\' Pi) \t\t {0:.4f}'.format(pi(dataset)))
    print('κ* (Fleiss\' Kappa) \t {0:.4f}'.format(kappa(dataset)))
    print('Artstein-Poesio Bias \t {0:.4f}'.format(bias(dataset)))
    
show_similarity_and_agreement(human_segmentations)
```

```python
def pi_variance(dataset):
    # This uses formula 13 in Fleiss et al. 1979 (single-category π* variance)
    n_coders = len(dataset.coders)
    n_possible_boundaries = potential_boundaries(dataset)
    return Decimal(2) / (n_possible_boundaries * n_coders * (n_coders-1))

def pi_z_test(dataset):
    # This uses formula 16 in Fleiss et al. 1979 (single-category π* z score)
    n_coders = len(dataset.coders)
    n_possible_boundaries = potential_boundaries(dataset)
    n_actual_boundaries = boundaries_placed(dataset)
    p_boundary = Decimal(n_actual_boundaries) / (n_possible_boundaries * n_coders)

    a = pi(dataset) + (Decimal(1)/(n_possible_boundaries*(n_coders-1)))
    b = ((n_possible_boundaries*n_coders*(n_coders-1)) / Decimal(2))**Decimal(0.5)
    
    return a * b

print('Var(π*) (single category) \t {0:.5f}'.format(pi_variance(human_segmentations)))
print('z-score (single category) \t {0:.4f}'.format(pi_z_test(human_segmentations)))
```

```python
def sort_annotator_names(s):
    parts = s.split(',')
    return ','.join([parts[0]] + sorted(parts[1:]))

def plot_pairwise_similarity_scores(dataset):
    scores = sorted([ (sort_annotator_names(label), score) for label, score 
                        in b(dataset).items() ])
    num_comparisons = len(list(it.combinations(dataset, 2)))
    pairs = [ ' vs. '.join(strip_prefixes(t[0].split(',')[1:])) 
                for t in scores[:num_comparisons] ]
    # Group pairwise similarity scores by interview.
    key = lambda v: v[0].split(',')[0]
    grouped = [ [k]+[t[1] for t in g] for k,g in it.groupby(scores, key) ]
    # Now we have: 
    # [ [interview A, coder 1 vs. coder 2, coder 1 vs. coder 3, coder 2 vs. coder 3],
    #   [interview B, coder 1 vs. coder 2, coder 1 vs. coder 3, coder 2 vs. coder 3] ... ]
    dot_plot(list(zip(*grouped))[1:], interview_ids, 'Pairwise segmentation similarity', pairs, minv=0.0)

null_segmentations = null_segmentations_for(human_segmentations)
random_segmentations = random_segmentations_for(human_segmentations)
regular_segmentations = regular_segmentations_for(human_segmentations)
texttiling_segmentations = load_dataset('texttiling-parameter-sweep/TextTiling-w16-k10-m+0.25-n1-s1.json')
topictiling_segmentations = load_dataset('topictiling/topictiling+2.00.json')
bayes_segmentations = load_dataset('segmentations-bayes-stem-stop-α0.189.json')
bayes_nnvbjj_segmentations = load_dataset('segmentations-bayes-nn-vb-jj-stem-α0.136.json')
bayes_corefs_segmentations = load_dataset('segmentations-bayes-corefs-stem-stop-α0.093.json')

humans_plus_null = merge_datasets(human_segmentations, null_segmentations)
humans_plus_random = merge_datasets(human_segmentations, random_segmentations)
humans_plus_regular = merge_datasets(human_segmentations, regular_segmentations)
humans_plus_texttiling = merge_datasets(human_segmentations, texttiling_segmentations)
humans_plus_topictiling = merge_datasets(human_segmentations, topictiling_segmentations)
humans_plus_bayes = merge_datasets(human_segmentations, bayes_segmentations)
humans_plus_bayes_nnvbjj = merge_datasets(human_segmentations, bayes_nnvbjj_segmentations)
humans_plus_bayes_corefs = merge_datasets(human_segmentations, bayes_corefs_segmentations)

for dataset in [ human_segmentations, 
                 humans_plus_bayes_nnvbjj,
                 humans_plus_texttiling, 
                 humans_plus_bayes_corefs,
                 humans_plus_bayes,
                 humans_plus_regular, 
                 humans_plus_random, 
                 humans_plus_null,
                 humans_plus_topictiling, 
               ]:
    plot_pairwise_similarity_scores(dataset)
    show_similarity_and_agreement(dataset)
```

**Number of boundaries placed**

Compare the number of boundaries placed by each coder. Rather than comparing raw boundary counts, look at the number of boundaries placed out of all possible boundaries. A value of 1 would indicate that a boundary was placed between every sentence; a value of 0 would indicate that no boundaries were placed.

```python
def divide_ratios(boundary_ratios):
    from operator import truediv
    return [ [ truediv(*boundary_ratios['interviews:%s'% item][coder]) for item in interview_ids ] 
             for coder in coders ]

dot_plot(divide_ratios(boundary_ratios_for(human_segmentations)), 
         interview_ids, 
         'Ratio of boundaries placed to possible boundaries', 
         strip_prefixes(coders))
```

**Evaluation of TextTiling**

Compare a hypothetical segmentation generated by texttiling to segmentations by individual coders.

```python
# Parameters:
# w: pseudosentence (token sequence) size, 14–26
# k: block size, 6–14
# m: depth score cutoff, -1.0–+1.0
#    -0.75 means a cutoff of (mean(depth) - 0.75*(stddev(depth))
#    +0.75 means a cutoff of (mean(depth) + 0.75*(stddev(depth))
# n: number of rounds of smoothing, 1–2
# s: smoothing width, 1–2
w = [ 'w%s' % i                for i in range(14, 27, 2) ]
k = [ 'k%s' % i                for i in range(6, 15, 2) ]
m = [ 'm{:+.2f}'.format(i/100) for i in range(25,101,25) ]
n = [ 'n%s' % i                for i in range(1, 4) ]
s = [ 's%s' % i                for i in range(1, 4) ]

param_combos = [ '-'.join(params) for params in it.product(w,k,m,n,s) ]

texttiling = { params: load_dataset('texttiling-parameter-sweep/TextTiling-%s.json' % params) 
               for params in param_combos}
```

Mean of B comparisons with individual coders over all items, and effect on inter-coder agreement (as drop in Fless' Pi).
